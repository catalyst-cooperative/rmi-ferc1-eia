```python
# Pandas for data analysis:
import pandas as pd

# SQLAlchemy gives us a bunch of database tools:
import sqlalchemy as sa

# Import the top level PUDL package:
import pudl
```

```python
pudl_settings = pudl.workspace.setup.get_defaults()
ferc1_engine = sa.create_engine(pudl_settings["ferc1_db"])
```

```python
f1_plant_in_srvc = pd.read_sql(
    """SELECT f1_respondent_id.respondent_name, f1_row_lit_tbl.row_literal, f1_plant_in_srvce.*
FROM f1_respondent_id INNER JOIN (f1_plant_in_srvce INNER JOIN f1_row_lit_tbl ON (f1_plant_in_srvce.report_year = f1_row_lit_tbl.report_year) AND (f1_plant_in_srvce.row_number = f1_row_lit_tbl.row_number) AND (f1_plant_in_srvce.row_seq = f1_row_lit_tbl.row_seq)) ON f1_respondent_id.respondent_id = f1_plant_in_srvce.respondent_id
WHERE (((f1_row_lit_tbl.sched_table_name)="f1_plant_in_srvce"))
ORDER BY f1_plant_in_srvce.report_year, f1_respondent_id.respondent_name, f1_plant_in_srvce.spplmnt_num, f1_plant_in_srvce.row_number;""", ferc1_engine)
```
